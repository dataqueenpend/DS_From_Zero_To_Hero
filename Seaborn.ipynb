**Using the distribution plot**

*Seaborn* supports customizing plots by adding arguments to the line of code. When you want to create a simple plot you can disable KDE and change the number of bins you want to use. 

```
sns.distplot(df['col1'], kde=False, bins=10)
```

Output will return histogram like the basic Pandas histogram.

**Alternative ways of data distribution**

As needed we can built a plot in rug style, which is an alternative way of showing data on the plot. If you want, you can use KDE in this plot.

```
sns.distplot(df['col1'], hist=False, rug=True)]
```

Trying different ways of showing your data can give you important insights. 


**Further customizations**

It is possible to further customize the plot by passing arguments to underlying functions, because distplot() relies on additional *Seaborn* function as a kdeplot() or rugplot(). As in below example, passing a kde_kws dictionary to the function kde() tells it to shade the plot.

```
sns.distplot(df['col1'], hist=False, rug=True, kde_kws={'shade':True})
```

**Regression plots**

The previous type of analysis we've performed are often called univariate analysis. It is because we're looking only on one variable. *Regression* analysis is reffered as bivariate, because there is two variables in the play.

**Regplot()**

*Regplot* is a basic function to build regression plots in Seaborn. It generates scatter plot with regression line. The data, x and y axis must be defined. But exept those, it's creation goes as in distplot().

```
sns.distplot(data=df, x='col1', y='col2')
```

Some confusion - while working with Seaborn - can raise a fact, that there are couple of ways to do the same plot. 

**lmplots() and regplots()**

Base to the lmplots() is a regplot(). Similar as in relation of kdeplot() and distplot(). Higher type of Seaborn plot - here the lmplot() - can produce more powerful output. But if we set up those plots in similar manner, the output will show almost no differences. Execept one - aspect ration. But there is much more to lmplots() than only this!

**lmplot() faceting**

Can be done by using both hue, or col arguments. Use *hue*, when you want to organize data by colors. And *col*, when you want to organize it by columns. 

```
# Create a regression plot using hue
sns.lmplot(data=df,
           y="premiums",
           x="insurance_losses",
           hue="Region")

# Show the results
plt.show()
```

**Regression plot with multiple rows**

```
# Create a regression plot with multiple rows
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           row="Region")

# Show the plot
plt.show()
```

**Theme examples with sns.set_style()**

Loop with different styles in *Seaborn* - choose which best suits you!

````python
for style in['white', 'dark', 'whitegrid', 'darkgrid', 'ticks']:
  sns.set_style(style)
  sns.distplot(df['year'])
  plt.show()
````

**Removing axes with despine()**

If you feel that some "junk chart" is making the plot harder to read use despine(). Common use of dispine() is to remove the axes from chart called spine. 

````python
sns.set_style('white')
sns.distplot(df['year'])
sns.despine(left=True)
````

By default despine() removes top and right lines, but it can be adjustable with passing additional arguments to the function.

**Colors with Seaborn**

Chosing right color pallette for the data visualization can lead to more effective interpretation of the data. It is also good practice to keep in mind, that 8% of the population is affected by color blindness, there are some pallettes that are tailored for those users.

There are couple of functions in *Seaborn* for creating, viewing and configuring color palettes. *Seaborn* - as we know already - is built on top of the *Matplotlib*. In this case that means, that it is able to interpret *Matplotlib* color codes. 

Below example of applying 'g' - green to Seaborn plot. 

````python
sns.set(color_codes='True')
sns.distplot(df['year'], color='g')
````

**Palettes**

Using colors seperately can be tricky! Therefore, often we use color palettes to change the appearance and style of the plot. And create your own your own custom sequential, categorical, or diverging palettes. 

````python
#cycling color palette throughout the plot
for p in sns.palettes.SEABORN_PALETTES:
  sns.set_palette(p)
  sns.distplot(df['year'])
````

**Displaying palettes**

*Seaborn* has some useful functions for chosing the right palette:
* sns.palplot() - function which display a palette
* sns.color_palette() - returns the current palette

````python
for p in sns.palettes.SEABORN_PALETTES:
  sns.set_palette(p)
  sns.palplot(sns.color_palette()) #returns color swatches
  plt.show()
````

**Defining custom palettes**

* Circular color palettes - when **categorical** data is not ordered

````python
sns.palplot(sns.color_palette('Paired', 12))
````

* Sequential palettes - when data has a consistent range from low to high values

````python
sns.palplot(sns.color_palette('Blues', 12))
````

* Diverging palettes - when both low and high values are interesting

````python
sns.palplot(sns.color_palette("BrBG", 12))
````
