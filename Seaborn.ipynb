**Distplot() function**

Using distplot function you can create plot similar to the histogram, but in reality is a KDE (Gaussian Kernel Density Estimate). 

```
import seaborn as sns
sns.distplot(ds['col1'])
```

**Pandas Histogram vs. Seaborn Distplot()**

* actual frequency of observations | * automatic labeling on x axis
* wide bins                        | * customizable color palette
* no automatic labels              | * narrow bins


Using *Seaborn* you can build powerful and informative plots. As an advantage for *Seaborn* goes also its customizations capabilities - for making visualization more pleasing and eye catching.

**Using the distribution plot**

*Seaborn* supports customizing plots by adding arguments to the line of code. When you want to create a simple plot you can disable KDE and change the number of bins you want to use. 

```
sns.distplot(df['col1'], kde=False, bins=10)
```

Output will return histogram like the basic Pandas histogram.

**Alternative ways of data distribution**

As needed we can built a plot in rug style, which is an alternative way of showing data on the plot. If you want, you can use KDE in this plot.

```
sns.distplot(df['col1'], hist=False, rug=True)]
```

Trying different ways of showing your data can give you important insights. 


**Further customizations**

It is possible to further customize the plot by passing arguments to underlying functions, because distplot() relies on additional *Seaborn* function as a kdeplot() or rugplot(). As in below example, passing a kde_kws dictionary to the function kde() tells it to shade the plot.

```
sns.distplot(df['col1'], hist=False, rug=True, kde_kws={'shade':True})
```

**Regression plots**

The previous type of analysis we've performed are often called univariate analysis. It is because we're looking only on one variable. *Regression* analysis is reffered as bivariate, because there is two variables in the play.

**Regplot()**

*Regplot* is a basic function to build regression plots in Seaborn. It generates scatter plot with regression line. The data, x and y axis must be defined. But exept those, it's creation goes as in distplot().

```

For taking advantage of underlying *Matplotlib* we need to create subplots and then pass the resulting axes object to the *Seaborn* function.

In above example, we add - by setting ax - xlabel. 

This is only and introduction to customizing the plot with *Matplotlib* axes object.

In above example we've changed xlabel, ylabel, title and the range of years showing on the plot (with parameter xlim=())

**Caterogical plot types**

Linear regression and distribution plots are great for plotting numerical data. But what about categorical (object type) data? *Seaborn* also can be used to create plots for this kind of data! What are examples of categorical data?
* geographical data (state, country, region)
* gender
* eye color
* blood type
* ethnicity

**3 groups of categorical data plots in Seaborn**

1. *stripplot()* and *swarmplot()* - shows all the individuals observations on the plot
2. *boxplot()*, *violinplot()*, *lvplot()* - shows abstract representation of the categorical data
3. *barplot()*, *countplot()*, *pointplot()* - shows statistical estimates of the categorical values

*Barplot()* and *countplot()* shows summaries of the data. *Countplot()* is used to for present number of instances of each observations. 

**Plots of each observations - stripplot()**

Each observation can be difficult to see while using stripplot(), but we can the jitter parameter to make it easier. 


````python
sns.stripplot(data=df, y='state', x='firespots', jitter=True)
````

**Plots of each observation - swarmplot()**

Swarmplot() use complex algorithm to place observations in the manner where they do not overlap. But it is not scaling well in large datasets (as you can see below).

````python
sns.swarmplot(data=df, y='state', x='firespots')
````

**Abstract representation of the data - boxplot()**

Boxplot() is used to show several information related to the data distribution, including: median, upper and lower quartiles, outliers. 

````python
sns.boxplot(data=df, y='state', x='firespots')
````

**Abstract representation - violinplot()**

Violinplot() is a combination of a kernel density plot and boxplot. Can be used as an alternative of providing alternative view of the data distribution. Kernel density calculation is great for large datasets (it does not show each observation), but can be computationally intensive to produce. 

````python
sns.violinplot(data=df, y='state', x='firespots')
````

**Abstract representation - lvplot()**

Lvplot() stands for letter - value plot. Is a hybrid between boxplot and violinplot. Scales more effectively for large datasets. Faster to render and easier to interpret. 


* According to official documentation lvplot() is longer available in Seaborn. https://seaborn.pydata.org/whatsnew.html

````python
sns.lvplot(data=df, y='state', x='firespots')
````

**Statistical estimates - barplot** 

Shows estimates of a values as well as confidence interval. 

````python
sns.barplot(data=df, y='state', x='firespots')
````

**Statistical estimates - pointplot**

Similar to the barplot(), but shows summary measure and confidence interval. Can be very useful for observing how values change across the categorical values.

````python
sns.pointplot(data=df, y='state', x='firespots', hue='state')
````

**Statistical estimates - countplot()**

Displays a number of instances in each variable.

````python
sns.countplot(data=df, x='firespots', hue='state')
````

**Matrix plots**

The most common type of matrix plot is a heatmap. Heatmap can be easily created using Seaborn. What use of a matrix plots is? To quickly spot the trends in the datatset. 

**Getting data in the right format** Seaborn heatmap() expects that data will be in a matrix. Here we can use pandas function *crosstab()* - which is comonly use to manipulate the data. 

````python
pd.crosstab(df['year'], df['month'], values=df['firespots'], aggfunc='mean')
````

*Crosstab* function summarizes the data on firespots in our Brazilian Rainforest df by the year and month. In above example we also used the aggfunc argument to get an average number of firespots in each year and month combinations.

**Building a heatmap** Heatmap translates numeric values in the matrix into a color coded grid.

````python
sns.heatmap(pd.crosstab(df['year'], df['month'], values=df['firespots'], aggfunc='mean'))
````

**Customizing the heatmap**

```
sns.heatmap(df_crosstab, annot=True, fmt='d', cmap='YlGnBu', cbar=False, linewidths=.5)
```

Annot equals True disable the annotations for individual cells. The format (fmt) parameter ensures that the results are displayed as integers. Cmap is for custom color formatting. Cbar attribute set to False ensures us that the color bar will not be displayed in the heatmap. Linewidths parameter with a small value will give a a little space between the cells. 

**Centering the heatmap** With Seaborn we can center the color scheme of a heatmap on specific value. In a example we center on a value store on Semptember and Saturday, using the numeric indices for these values.

```
sns.heatmap(df_crosstab, annot=True, fmt='d', cmap='YlGnBu', cbar=False, linewidths=.5, center=df_crosstab.loc[9,6])
```

This change results that the color scheme is shifted towards yellows instead of blues.

**Plotting a correlation matrix**
One of the common uses of a heatmap() is to visually represent correlation between the variables. We do it with pandas corr() function, which is calculating correlation between the values in the columns. 

````python
sns.heatmap(df.corr())
````

This correlation can be useful to undestand which variables you might want to study further using regplot().

Create a heatmap of the data using the BuGn palette
Disable the cbar and increase the linewidth to 0.3

```
# Create the crosstab DataFrame
pd_crosstab = pd.crosstab(df["Group"], df["YEAR"])

# Plot a heatmap of the table with no color bar and using the BuGn palette
sns.heatmap(pd_crosstab, cbar=False, cmap="BuGn", linewidths=.3)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

#Show the plot
plt.show()
plt.clf()
```
