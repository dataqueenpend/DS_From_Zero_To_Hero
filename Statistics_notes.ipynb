##The Cumulative Distribution Function - CDF

image.png

denoted by capital F

And it is defined simply as the probability that the random variable, X, is less than or equal to some value, little x.

* the CDF is always going to be non-decreasing in x
* always between 0 and 1
* the limit, as x goes to negative infinity, of the CDF is equal to 0; And, along with that is the limit, as x goes to positive infinity of the CDF, is equal to 1.
