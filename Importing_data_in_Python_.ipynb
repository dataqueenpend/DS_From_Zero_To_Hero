If in the flatfile there are some rows, that you don't want to import (such as header), maybe specific columns that you don't need. Or maybe your flatfile uses different delimiter than a comma.

### Alternative way to deal with importing different data types

Np.loadtxt() can be hard to work with when you will be dealing with multiple data types while importing data in Python. Another way to import files with multiple data types is to do so by using **np.genfromtxt()** . Passing to it **dtype=None** will make it to figure out what kind of data is importing to the shell.

```
data = np.genfromtxt('titanic.csv', delimiter=',', names=True, dtype=None)
```

First argument of the function is csv file name, second one is delimiter, and the third one is telling the function, that we have a header here.

Flat files imported with this function are stored as a 1d structed numpy arrays. This solves an issue with diffent data types - in numpy arrays all the data types are unified. **Why 1d numpy array?** In 1d numpy array each element of the array is a row from the flat file imported. 

Shape of the array we can check by passing **np.shape(var_name)**. How to access rows and columns of the array? Do it by just typing the name of the array and the name of the col/row you want to access. 

```
data[col_or_row_name]
```

## Importing HDF5 files

HDF5 files popularity is growing and will be growing as a standard for mechanism for large quantities of numerical data. How large **Hierarchical Data Format** type files can handle? Hundreds of gigabytes, terrabytes? No. It can do more. **Exabytes**. 

To use HDF5 files, first we need to import package *h5py*. Then assign the hdf5 file to a variable and load it into the environment. Remember to set 'r' - for read only mode. 

```
import h5py
filename = 'H-H1_LOSC_4_V1-815411200-4096.hdf5'
data = h5py.File(filename, 'r') # 'r' is to read
```

Exploring HDF5 file is somewhat similar to exploring a Python dictionary. We do it with the key method. 

```
for key in data.keys():
print(key)
meta
quality
strain
```

As you see there are 3 keys. Every one of them is an HDF group. We can think of keys as a directories for this groups. LIGO documentation tells us a little more about what they are:
* **Meta** - for metadata of the file
* **Quality** - with informations about data quality
* **Strain** - 'strain data from the interferometer, the main measurement performed by LIGO, the data of interest'.


**How to find what type of data is store in particular directory?** 

```
for key in data['meta'].keys():
print(key)
Description
DescriptionURL
Detector
Duration
GPSstart
Observatory
Type
UTCstart
```


When you know the keys in metadata dictionary, you can **access any metadata** you want. All you need to do is just print it **using .value method**. 

```
print(data['meta']['Description'].value, data['meta']['Detector'].value)
b'Strain data time series from LIGO' b'H1'
```

## Importing MATLAB files

**MATLAB** stands for Matrix Laboratory, is a standard for engineering and science for completing various numerical computing tasks. It is because of *MATLAB*'s powerful linear algebra and matrix capabilities. The extension for *MATLAB* files is simply ```.mat```.

*MATLAB* file can contain multiple object types - such as strings, arrays, floats and much more. And the *MATLAB* file is a collection of all these object store in variables. Loading *MATLAB* into working environment we should expect a lot of different variables and objects.

In order to import *MATLAB* files to the Python enviroment we need to use another Python library called **SciPy**. There are 2 functions that we will use to perform it:

```

### Exploiting table relationships

To this time we've covered how to import records from single table in relational database. But, as we know already, the true power of relational databases is, that they are related to eachother. 

Going down this road, you can predict, that it should be some way, that we can import interesting to us records from multiple table in same relational database. 

Tables related by some particular feature (column name and the same unique input in the specific column in all related tables) can be joined in Python with Pandas. 

**Joining tables in Python** 

*Inner Join*

```
from sqlalchemy import create_engine
import pandas as pd
engine = create_engine('sqlite:///database.sqlite')
df = pd.read_sql_query("SELECT col1_table1, col2_table2 FROM table1
INNER JOIN table2 on table1.col2_table1 = table2.col1_table2", engine)
print(df.head())
```
